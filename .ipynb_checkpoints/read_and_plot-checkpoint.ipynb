В данном разделе будут рассмотрены подходы к загрузке, первичной обработке и визуализации исходных данных.

```python
# Библиотеки (пакеты)
import os                       # Системные функции (работа с директориями, файлами и т.п.)
import shutil                   # unix-подобные команды
import patoolib                 # Работа с архивами (7z, rar)
import numpy as np              # Работа с массивами (матрицами)
import matplotlib.pyplot as plt # Визуализация
import cv2                      # OpenCV - обработка изображений
import cartopy                  # Построение карт
```

```python
# Создаём временную папку для распаковки архива. Там будет храниться его содержимое.

tmp_dir = 'tmp'

if os.path.exists(tmp_dir):
    shutil.rmtree(tmp_dir)
    
os.mkdir(tmp_dir)

os.listdir('tmp')
```

```python
# Выбираем первый файл из директории "data" для примера
filename = os.path.join('data', os.listdir('data')[0])

# Распаковываем архив
patoolib.extract_archive(filename, outdir=tmp_dir, verbosity=-1, interactive=False)
```

```python
# Откроем первый файл
with open('tmp/201906111200_IR_039', 'rb') as file:
    first = int.from_bytes(file.read(4), byteorder='little') # Начало области (не используется)
    second = int.from_bytes(file.read(4), byteorder='little') # Конец области (не используется)
    pixs = int.from_bytes(file.read(4), byteorder='little') # Количество пикселей в строке
    rows = int.from_bytes(file.read(4), byteorder='little') # Количество строк
    
    print('first: {}, second: {}, pixs: {}, rows: {}'.format(first, second, pixs, rows))
    
    # Преобразуем массив байт в числовой массив
    img = np.frombuffer(file.read(rows * pixs * 2), dtype=np.uint16)
    # Преобразуем размерность
    img = np.reshape(img, (rows, pixs))
    # Лучше заранее выполнить преобразование типов для дальнейших расчётов
    img = img.astype(np.float32) / 10
```

```python
# Визуализируем данные
plt.imshow(img, cmap='gray')
```

```python
"""
Набор функций для получения массива с изображениями
"""

# Удобная функция для чтения одного канала
def read_img(filename):
    with open(filename, 'rb') as file:
        first = int.from_bytes(file.read(4), byteorder='little')
        second = int.from_bytes(file.read(4), byteorder='little')
        pixs = int.from_bytes(file.read(4), byteorder='little')
        rows = int.from_bytes(file.read(4), byteorder='little')
        img = np.frombuffer(file.read(rows * pixs * 2), dtype=np.uint16)
        img = np.reshape(img, (rows, pixs))
        img = img.astype(np.float32)
    return img

# Удобная функция для чтения всех каналов из архива. Возвращает словарь (ассоциативный массив)
def read_images(filename):
    img_dict = {}
    
    tmp_dir = 'tmp'

    if os.path.exists(tmp_dir):
        shutil.rmtree(tmp_dir)

    os.mkdir(tmp_dir)

    # Распаковываем архив
    patoolib.extract_archive(filename, outdir=tmp_dir, verbosity=-1, interactive=False)
    
    # Выбираем строку периода из пути к файлу (tmp/202006191200.rar)
    period = os.path.basename(filename).split('.')[0]
    
    for chan_name in ['IR_108', 'WV_062', 'WV_073', 'VIS006']:
        img_dict[chan_name] = read_img('{}/{}_{}'.format(tmp_dir, period, chan_name))
    
    return img_dict
```

```python
img_dict = read_images('data/201906111200.rar')

img_dict.keys()
```

```python
# Визуализируем два канала
fig, ax = plt.subplots(1, 2)
ax[0].imshow(img_dict['IR_108'], cmap='gray')
ax[0].title.set_text('IR_108')
ax[1].imshow(img_dict['WV_062'], cmap='gray')
ax[1].title.set_text('WV_062')
```

```python
# Рассчитаем разность каналов
btd_1 = img_dict['WV_062'] - img_dict['IR_108']

# Визуализируем результат
plt.imshow(btd_1, cmap='gray')
```

```python
# Маска "заднего плана" - за пределами Земного диска
bg_mask = img_dict['IR_108'] == 0

# Заполняем чёрным цветом область за пределами Земного диска 
btd_1[bg_mask] == 0

# Визуализируем результат
plt.imshow(btd_1, cmap='gray')

# Ничего не изменилось, так как в массиве присутствуют отрицательные значения!
```

```python
# Нормализируем значения разностей для корректной визуализации

# Рассчитаем разность каналов
btd_1 = img_dict['WV_062'] - img_dict['IR_108']

btd_1_norm = np.zeros_like(btd_1)

# Границы нормализации
min_val, max_val = -1000, 20

# Учитываем только значения в пределах Земного диска
values = btd_1[~bg_mask].copy()

# Уберём значения вне границ (приравняем к ближайшей границе)
values = np.clip(values, min_val, max_val)

# Нормализация 
values = np.round((values - min_val) / (max_val - min_val) * 255)

btd_1_norm[~bg_mask] = values

# Меняем тип для 255-битного изображения
btd_1_norm = btd_1_norm.astype(np.uint8)

plt.imshow(btd_1_norm, cmap='gray', vmin=0, vmax=255)
```

```python
# Сохраним результат в формате PNG

if not os.path.exists('results'):
    os.mkdir('results')
    
cv2.imwrite('results/btd_1.png', btd_1_norm)
```

```python
# Сохраним результат в формате BIN

if not os.path.exists('results'):
    os.mkdir('results')
    
with open('results/btd_1.bin', 'wb') as file:
    btd_1.astype(np.int32).tofile(file)
```

```python
# Прочитаем результат в формате BIN
    
with open('results/btd_1.bin', 'rb') as file:
    btd_1_bin = np.fromfile(file, dtype=np.int32)
    btd_1_bin = np.reshape(btd_1_bin, btd_1.shape)
    
plt.imshow(btd_1_bin, cmap='gray')
```
